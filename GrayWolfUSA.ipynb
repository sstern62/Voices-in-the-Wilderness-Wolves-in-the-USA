```python
text = []

for page in reader.pages:
    text.append(page.extract_text())
```

```python
text[1]
```

```python
joined_text = ' '.join(text)
```

```python
with open('graywolf.txt', 'w', encoding='utf-8') as f:
    f.write(joined_text)
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.text import Text
nltk.download('stopwords')
```

```python
tokenized = nltk.word_tokenize(joined_text)
```

```python
tokenized[:10]
```

```python
new_text = Text(tokenized) #capital "T" Text passes allows the data to be analyized by NLTK
```
